# C2 Program
Explore the automobile dataset and visualize the i) Distribution of the two and four door cars with respect to the type of fuel they use ii) Distribution of cars of different body styles with respect to the type of fuel they use ii) Total number of each type of body style cars categorized by fuel type iii) Horsepower of each of the fuel type with reference to the type of drive wheel present in cars. (https://www.kaggle.com/datasets/toramky/automobile-dataset?resource=download)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Automobile_data.csv')
df.head()
```

```python
sns.countplot(x = 'num-of-doors',hue = 'fuel-type',data = df)
plt.title("Distribution of the two and four door cars w.r.t the type of fuel they use")
plt.show
```

```python
sns.countplot(x = 'body-style',hue = 'fuel-type',data = df)
plt.title("Distribution of cars of different body styles with respect to the type of fuel they use")
plt.show
```

```python
sns.countplot(x = 'body-style',hue = 'fuel-type',data = df)
plt.title("Total number of each type of body style cars categorized by fuel type")
plt.show
```

```python
df.replace('?',np.nan,inplace=True)
df = df.explode('horsepower')
df['horsepower'] = df['horsepower'].astype(float)
sns.boxplot(data=df, x='fuel-type', y='horsepower', hue='drive-wheels')
plt.show()
```
